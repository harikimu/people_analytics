## Linear Regression Modelling

```python
import pandas as pd
df= pd.read_excel("C:\\Users\\heeckim\\Desktop\\work\\12_python\\pyharry\\resources\\lrdata.xlsx")
```

```python
import statsmodels.api as sm
X = df['performance_rating']
Y = df['percentage_salary_hike']
X = sm.add_constant(X)
model1=sm.OLS(Y,X) # Ordinary Least Square Model  사용
result=model1.fit()
result.summary()
```

* const/coef = -9.5358 = performance_rating이 0일때의 추정치
* performance_raiting/coef = 7.8463 = 슬로프, 퍼포먼스레이팅이 1증가할때마다 셀러리 7.8463 증가
* p(>|t|): 슬로프의 유의미성, 0.05보다 작으면 통계적으로 유의미하다
* salary hike에 대하여 performance rating로 예측하는 회귀분석을 실시한 결과, 이 회귀모형은 통계적으로 유의미하였다(F(Df Model,Df Residual) = F-statistics, p < 0.05) = (F(1, 1468) = 2187, p < 0.05>)
* performance rating의 slpoe는 7.8463으로 salary hike에 유의미한 예측 변인인 것으로 나타났다(t(1468)=46.767, p < 0.05)
