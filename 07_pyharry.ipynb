# Covariance, Correlation

```python
df.age.cov(df.salary_2018) # age와 salary_2018 코바리언스 반환, 양수이면 연관성이 있다고 보지만 이 샘플 데이터의 정확도는 높지않다.
```

```python
df1=df[['age', 'salary_2018', 'totalW_working_years', 'percent_hike']]
df1.cov() # df에서 선택한 4개의 컬럼의 코바리언스를 테이블(매트릭스)로 반환
```

```python
df.age.corr(df.salary_2018) # age와 salary_2018 코릴레이션 반환, 1 또는 -1에서 가까울수록 연관성이 높다. 0.5 미만은 연관성이 약함.
```

```python
df1.corr() # df에서 선택한 4개의 컬럼의 코릴레이션을 테이블(매트릭스)로 반환
```
