```python
import pandas as pd
df = pd.read_excel('C:\\Users\\heeckim\\data_analytics\\pyharry\\resources\\lrdata.xlsx')
```

```python
df1 = df.drop(['percentage_salary_hike'], axis=1) # salary hike를 제외한 모든 컬럼을 선택
```

```python
import statsmodels.api as sm
x = df1
y = df['percentage_salary_hike']
x = sm.add_constant(x) # Add a column of ones to an array.
model1=sm.OLS(y,x).fit() # Ordinary Least Squares
model1.summary()
```

```python
# p값이 0.05 이하인 경우 H0을 기각후 H1을 채택, 위와같은 경우 Performacne Ratiing 0임으로, HO 관계가 없다를 기각하고 H1 연관성이 있다를 선택
```
