# Cross-Tab, Proportion Table, Stacked/Bar/Column Chart, Mosaic Plot


```python
import pandas as pd
df = pd.read_excel('C:\\Users\\heeckim\\Desktop\\work\\12_python\\pyharry\\resources\\mean median mode.xlsx')
```

```python
pd.crosstab(df.department,df.gender) # CrossTab을 반환
```

```python
pd.crosstab(df.department,df.gender,normalize=True) # Proportion Table을 반환 (100퍼센트 기준값)
```

```python
pd.crosstab(df.department,df.gender,normalize='index') # 로우로 그룹핑하여 Proportion Table을 반환 
```

```python
pd.crosstab(df.department,df.gender,normalize='columns') # 컬럼으로 그룹핑하여 Proportion Table을 반환
```

```python
pd.crosstab(df.department,df.gender).plot.bar(stacked=True) # 각 부서별 성별을 스택하여 스택바차트로 반환
```

```python
pd.crosstab(df.gender,df.department).plot.bar(stacked=True)
```

```python
pd.crosstab(df.gender,df.department, normalize='index').plot.bar(stacked=True) # 100퍼센트 기준의 스택바 차트를 반환
```

```python
pd.crosstab(df.department, df.gender, normalize='index').plot.bar(stacked=True)
```

```python
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df,['department','gender'])
plt.show()
```

```python
mosaic(df,['department','gender','over_time'])
plt.show()
```
